```python
import pandas as pd
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv') #Read the dataset
```

The dataset contains the ff facts:
- 41 columns (period/ date included)
- 5699 rows (header included)
- 38 columns have the string data type 
- 3 columns have float data type
- dataset contains some null values

Owing to the above facts, the dataset require data cleaning:
- rename the columns 'Period\Unit' and '[US dollar]' to 'Time' and 'US_dollar' respectively
- change 'Time' column to datetime datatype
- sort the values by 'Time' in ascending order
- rest index and drop initial index

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)     #rename the columns 'Period\Unit' and '[US dollar]' to 'Time' and 'US_dollar' respectively
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])   #change 'Time' column to datetime datatype
exchange_rates.sort_values('Time', inplace=True)     #sort the values by 'Time' in ascending order
exchange_rates.reset_index(drop=True, inplace=True)    #rest index and drop initial index
exchange_rates.head()
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()   #isolate the 'Time' and 'US_dollar' columns to a variable
euro_to_dollar['US_dollar'].value_counts()
```

The 'US_dollar' column contains 62 rows with '-' characters. Below, we will drop all the rows with such characters as they will not be relevant to our analysis

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']   #drop all rows with '-' character in 'US_dollar' column
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)   #convert 'US_dollar' column to a float data type
euro_to_dollar.info()   #exploring 'euro_to_dollar' dataset to confirm changes
```

```python
### import matplotlib
import matplotlib.pyplot as plt

### required to display plot
%matplotlib inline 

### Adding plot
plt.plot(euro_to_dollar['Time'],
        euro_to_dollar['US_dollar'])
plt.show()
```

```python
### Adding various rolling mean plots
plt.figure(figsize=(9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Original values', weight='bold')

for i, rolling_mean in zip([2, 3, 4, 5, 6],
                           [7, 30, 50, 100, 365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'],
             euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title('Rolling Window: ' + str(rolling_mean), weight='bold')
    
### Auto-adjusts the padding between subplots
plt.tight_layout() 
plt.show()
```

```python
### Explore data for monthly(30day) rolling mean
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

## Data Visualization


*  The changes in euro-dollar rate during the coronavirus pandemic will be visualized. Data from the year 2020 will be used and the 2016-2019 data used as a baseline. 

*  A line plot was used to visualize the changes in the currencies during the two different periods.



```python
### isolate 2020 data
covid_pandemic = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year == 2020)] 
### isolate 2016-2019 data as baseline
pre_covid = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2016
                   ) & (euro_to_dollar['Time'].dt.year <= 2019)] 
```

```python
### Import FiveThirtyEight graphs
import matplotlib.style as style
style.use('fivethirtyeight')

### Add plots
fig,ax = plt.subplots(figsize=(8,3))
ax.plot(covid_pandemic['Time'],
        covid_pandemic['rolling_mean'],
        linewidth=3, color='#e23d28')

ax.plot(pre_covid['Time'],
        pre_covid['rolling_mean'],
        linewidth=1, color='#A6D785')

### Add separate tick labels
ax.set_xticklabels([])
ax.set_yticklabels([])

x = 735910.0
for year in ['2016', '2017', '2018', '2019', '2020','2021']:
    ax.text(x, 1.02, year, alpha=0.5, fontsize=12)
    x += 365
    
y = 1.045   
for rate in ['1.05', '1.10', '1.15', '1.20']:
    ax.text(735759.0, y, rate, alpha=0.5, fontsize=12)
    y += 0.05


### Add Title and Subtitle
ax.text(735759.0, 1.27, "Euro-USD rate dropped below 1.10 for the first time in 2 years",
        weight='bold')
ax.text(735759.0, 1.25, 'Euro-USD exchange rates between 2016 and 2020',
        size=12)

### Add Signature
ax.text(735759.0, 1.0, '©DATAQUEST' + ' '*98 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=10)

#Adjust grid transparency
ax.grid(alpha=0.5)

plt.show()
```
